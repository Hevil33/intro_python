```python
# this is a code cell
import numpy as np
import time
import matplotlib.pyplot as plt
```

Simulate long running time

```python
x = np.linspace(0, 2 * np.pi, 1000)
y = np.sin(x)
time.sleep(6)
```

```python
fig, ax = plt.subplots(dpi=100)
ax.plot(x, y)
ax.set_title(r"$\sin(x)$")
ax.set_xlabel("x")
ax.set_ylabel("y")
```
